```python
#removing snow_depth_mm since it doesn't impact the result
# Drop columns that are not important based on analysis
weather_data_reduced = weather_data.drop(columns='snow_depth_mm')
```

```python
weather_data_reduced.head()
```

Filling missing values

```python
# Define a function to fill missing values with city-wise mean
def fill_city_wise_mean(df, column_name, group_column):
    city_means = df.groupby(group_column)[column_name].transform('mean')
    df[column_name].fillna(city_means, inplace=True)

# Fill missing values in 'avg_temp_c' with city-wise mean
fill_city_wise_mean(weather_data_reduced, 'avg_temp_c', 'city_id')

# Fill missing values in 'min_temp_c' with city-wise mean
fill_city_wise_mean(weather_data_reduced, 'min_temp_c', 'city_id')

# Fill missing values in 'max_temp_c' with city-wise mean
fill_city_wise_mean(weather_data_reduced, 'max_temp_c', 'city_id')

# Fill missing values in 'precipitation_mm' with city-wise mean
fill_city_wise_mean(weather_data_reduced, 'precipitation_mm', 'city_id')

# Fill missing values in 'avg_wind_dir_deg' with city-wise mean
fill_city_wise_mean(weather_data_reduced, 'avg_wind_dir_deg', 'city_id')

# Fill missing values in 'avg_wind_speed_kmh' with city-wise mean
fill_city_wise_mean(weather_data_reduced, 'avg_wind_speed_kmh', 'city_id')

# Verify that there are no more missing values in 'avg_temp_c'
print(weather_data_reduced.isnull().sum())
```
